```python
%matplotlib inline
import random
import math

#Distancia euclidiana
def eucldist(p0,p1):
    dist = 0.0
    for i in range(0,len(p0)):
        dist += (p0[i] - p1[i])**2
    return math.sqrt(dist)
    
#Algoritmo K-Means
def kmeans(k,datapoints):

    # d - Mimenção dos dados
    d = len(datapoints[0]) 
    
    #Limitação de iterações para calculo dos centroides
    Max_Iterations = 50
    i = 0

    cluster = [0] * len(datapoints)
    prev_cluster = [-1] * len(datapoints)
    
    #Definição randômica para os centroides dos clusters
    cluster_centers = []
    for i in range(0,k):
        new_cluster = []
        cluster_centers += [random.choice(datapoints)]
        force_recalculation = False
        
    while (i < Max_Iterations)  :
        prev_cluster = list(cluster)
        force_recalculation = False
        i += 1
    
        #Atualiza os pontos do cluster
        for p in range(0,len(datapoints)):
            min_dist = float("inf")
            
            #Verifica a distância de todos os centroides
            for c in range(0,len(cluster_centers)):
                
                dist = eucldist(datapoints[p],cluster_centers[c])
                
                if (dist < min_dist):
                    min_dist = dist  
                    cluster[p] = c   
        
        #Atualiza a posição do cluster 
        for k in range(0,len(cluster_centers)):
            new_center = [0] * d
            members = 0
            for p in range(0,len(datapoints)):
                if (cluster[p] == k):
                    for j in range(0,d):
                        new_center[j] += datapoints[p][j]
                    members += 1
            
            for j in range(0,d):
                if members != 0:
                    new_center[j] = new_center[j] / float(members) 
                
                else: 
                    new_center = random.choice(datapoints)
                    force_recalculation = True
                    
            cluster_centers[k] = new_center
    
        
    return cluster, cluster_centers


alocacoes, clusters = kmeans(k,datapoints.T)
    
```
