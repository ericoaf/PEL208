```python
import numpy as np
 
class Perceptron(object):
    """Implements a perceptron network"""
    def __init__(self, input_size, lr=0.1, epochs=10):
        self.W = np.zeros(input_size+1)
        # add one for bias
        self.epochs = epochs
        self.lr = lr
    
    def activation_fn(self, x):
        #return (x >= 0).astype(np.float32)
        return 1 if x >= 0 else 0
 
    def predict(self, x):
        x = np.insert(x, 0, 1)
        z = self.W.T.dot(x)
        a = self.activation_fn(z)
        return x, a
 
    def fit(self, X, d):
        for _ in range(self.epochs):
            for i in range(d.shape[0]):
                x, y = self.predict(X[i])
                e = d[i] - y
                print(e)
                self.W = self.W + self.lr * e * x
```

```python
oAnd = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]])
d = np.array([0, 0, 0, 1])

perceptron = Perceptron(input_size=2)
perceptron.fit(oAnd, d)
print(perceptron.W)
```

```python
print(perceptron.predict(oAnd[0]))
print(perceptron.predict(oAnd[1]))
print(perceptron.predict(oAnd[2]))
print(perceptron.predict(oAnd[3]))
```
