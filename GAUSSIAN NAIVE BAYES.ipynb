```python
# leitura de dados

dados = pd.DataFrame(iris[:,[0, 1, 2, 3]]).astype(float)
dados.columns = ['sepalLength', 'sepalWidth', 'petalLength', 'petalWidth']
dados['Classe'] = iris[:, 4]
dados_previsao = dados

# treinamento
tabela = dados.groupby("Classe").agg([np.mean, np.std])

# funcao de probabilidade
def prob(x, d, m):
    return math.pow(1/math.sqrt(2*math.pi*d), (x-m)**2/2*(d**2))
```

```python
# previsao

# salva respostas
respostas = []

# para cada exemplo de teste
for indice, linha in dados_previsao.iterrows():
    
    # salva probabilidades
    probabilidades = []

    # para cada classe
    for classe in dados.Classe.unique():
        
        # para cada caracteristica
        for column in ['sepalLength', 'sepalWidth', 'petalLength', 'petalWidth']:
                # pega desvio padrao e media do treino
                d = tabela.loc[classe][column]['std']
                m = tabela.loc[classe][column]['mean']

                # pega X do teste
                x = linha[column]

                # calcula probabilidade e salva
                p = prob(x, d, m)
                probabilidades.append((classe, p))
#                 print("Class: {0} Caract: {1} - prob({2:.2f}, {3:.2f}, {4:.2f}) = {5:.2f}".format(classe, column, x, d, m, p))
    
    probabilidades2 = pd.DataFrame(probabilidades)
    multiplicacao_probabilidades = probabilidades2.groupby(0).prod()
    multiplicacao_probabilidades_classe = multiplicacao_probabilidades.join(probTipoFlor)
    multiplicacao_probabilidades_classe2 = multiplicacao_probabilidades_classe.prod(axis=1)
    resposta = multiplicacao_probabilidades_classe2.idxmax()
    respostas.append(resposta)
    
#     print(multiplicacao_probabilidades)
#     print(multiplicacao_probabilidades_classe)
#     print(multiplicacao_probabilidades_classe2)

#     print('')
#     print('Resposta:', resposta)
```

```python
# analisa resultados
dados["respostas"] = respostas
dados["Acertou?"] = dados.Classe == dados.respostas
dados["Acertou?"].value_counts()
```
